```python
import pandas as pd

dtr2015 = pd.read_csv('daily-treasury-rates.csv')
dtr2016 = pd.read_csv('daily-treasury-rates (1).csv')
dtr2017 = pd.read_csv('daily-treasury-rates (2).csv')
dtr2018 = pd.read_csv('daily-treasury-rates (3).csv')
dtr2019 = pd.read_csv('daily-treasury-rates (4).csv')
```

```python
dtr2015 = dtr2015[::-1]
dtr2016 = dtr2016[::-1]
dtr2017 = dtr2017[::-1]
dtr2018 = dtr2018[::-1]
dtr2019 = dtr2019[::-1]
```

```python
risk_free_rate = pd.concat([dtr2015, dtr2016, dtr2017, dtr2018, dtr2019])[['Date', '13 WEEKS COUPON EQUIVALENT']]
```

```python
risk_free_rate = risk_free_rate.rename(columns = {'13 WEEKS COUPON EQUIVALENT': 'rf_rate'})
```

```python
risk_free_rate['Date'] = pd.to_datetime(risk_free_rate['Date'])
```

```python
import matplotlib.pyplot as plt
plt.plot(risk_free_rate['Date'], risk_free_rate['rf_rate'])
```

```python
#save to csv:

risk_free_rate.to_csv('risk_free_rate.csv')
```
